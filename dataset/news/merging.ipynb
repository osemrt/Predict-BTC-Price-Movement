```python
import pandas as pd
import numpy as np
```

```python
filename1='ccn.com [all_currencies] [no_time].csv'
filename2='coindesk.com [all_currencies] [time].csv'
filename3='cointelegraph.com [only_btc] [no_time].csv'
filename4='mix.csv [only_btc] [time].csv'
```

```python
df_ccn=pd.read_csv(filename1)
df_coindesk=pd.read_csv(filename2)
df_cointelegraph=pd.read_csv(filename3)
df_mix=pd.read_csv(filename4)
```

# Sets only datetime

```python
df_coindesk.head()
```

```python
df_mix.head()
```

```python
# Concatenate the dataframes
df_datetime = pd.concat([df_coindesk, df_mix])
df_datetime.head()
```

```python
df_datetime.shape
```

```python
# Remove the duplicate headlines
df_datetime = df_datetime.drop_duplicates('headline', keep='last')
df_datetime.shape
```

```python
# Sort the dataframe by datetime
df_datetime = df_datetime.sort_values(by=['datetime'], ascending=False, na_position='first')
df_datetime.head(10)
```

```python
# Write the dataframe to a csv file
df_datetime.to_csv('news [datetime].csv', index=False)
```
